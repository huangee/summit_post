## Topic Modeling

```python
import pandas as pd
import pickle

data = pd.read_pickle("dtm_combined.pkl")
data
```

```python
# If not installed: conda install -c conda-forge gensim

from gensim import matutils, models
import scipy.sparse
```

```python
tdm = data.transpose() # to change document-term matrix to a term-document matrix for gensim
tdm.head()
```

```python
sparse_counts = scipy.sparse.csr_matrix(tdm)
corpus = matutils.Sparse2Corpus(sparse_counts)
```

```python
cv = pickle.load(open("cv_stop.pkl", "rb"))
id2word = dict((v,k) for k,v in cv.vocabulary_.items())
```

```python
lda = models.LdaModel(corpus=corpus, id2word=id2word, num_topics=2, passes=10)
lda.print_topics()
```
